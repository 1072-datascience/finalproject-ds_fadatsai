```python
# import package
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor
from sklearn import neighbors
```

```python
#  set data_path
dir_data = '../data/'
f = os.path.join(dir_data, 'BRAZIL_CITIES_final.csv')

# read file
data = pd.read_csv(f)
data.head()
```

```python
# check if there's any NaN
data.isnull().any(axis=0)
```

```python
# modeling
feature = list(data.columns)
x_train, x_test, y_train, y_test = train_test_split(data[feature], data['IDHM_y'], test_size=0.3, random_state=0)
```

```python
# Bagging
model_BaggingRegressor = BaggingRegressor()
model_BaggingRegressor.fit(x_train, y_train)
```

```python
y_pred = model_BaggingRegressor.predict(x_test)
```

```python
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print("R2: ", r2)
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
```

```python
# kNN
model_KNeighborsRegressor = neighbors.KNeighborsRegressor()
model_KNeighborsRegressor.fit(x_train, y_train)
```

```python
y_pred = model_KNeighborsRegressor.predict(x_test)
```

```python
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print("R2: ", r2)
```
